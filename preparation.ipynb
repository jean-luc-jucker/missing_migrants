## 1.1 Merge

```python
# Merge dataframes into one
dfs = [y2014, y2015, y2016, y2017, y2018, y2019, y2020, y2021]
df = pd.concat(dfs)
print('Shape ::::::::::::::::::::')
print(df.shape)
print('\nInfo ::::::::::::::::::::')
print(df.info())
print('\nHead ::::::::::::::::::::')
df.head()
```

# 2. Clean

## 2.1 Convert data types

```python
# Convert 'Total', currently a string, to integer

df['Total'] = df['Total'].astype(str) # it is important to first define it as string, since it is currently mixed
df['Total'] = df['Total'].str.replace(",","") # remove all commas
df['Total'] = df['Total'].astype(int) # Convert to int

# Convert all other numeric columns, currently float, to int
to_int = ['Quality','Dead', 'Missing', 'Females', 'Males', 'Children', 'Survivors']
df[to_int] = df[to_int].apply(pd.to_numeric).fillna(0).astype(int)

# Recheck data types
df.info()
```
