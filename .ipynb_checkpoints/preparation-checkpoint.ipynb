## 1.1 Merge

```python
# Merge dataframes into one
dfs = [y2014, y2015, y2016, y2017, y2018, y2019, y2020, y2021]
df = pd.concat(dfs, ignore_index=True) # we ignore index so that we avoid duplicated indexes
print('Shape ::::::::::::::::::::')
print(df.shape)
print('\nInfo ::::::::::::::::::::')
print(df.info())
print('\nHead ::::::::::::::::::::')
df.head()
```

# 2. Clean

## 2.1 Convert data types

```python
# Convert 'Total', currently a string, to integer

df['Total'] = df['Total'].astype(str) # it is important to first define it as string, since it is currently mixed
df['Total'] = df['Total'].str.replace(",","") # remove all commas
df['Total'] = df['Total'].astype(int) # Convert to int

# Convert all other numeric columns, currently float, to int
to_int = ['Quality','Dead', 'Missing', 'Females', 'Males', 'Children', 'Survivors']
df[to_int] = df[to_int].apply(pd.to_numeric).fillna(0).astype(int)

# Recheck data types
df.info()
```

## 2.2 Assess NAs

```python
df.isna().sum()
```

```python
# Print summary
print('NAs summary ::::::::::::::::::::')
print(df.isna().sum())

# Month ##########
  # Since source, here: https://archive.ph/oNzH4 is dated 1 December 2021,
  # and since value in 'Incident Date' is in December, we fill NA will 'December'
df.loc[df['Month'].isna(), 'Month'] = df.loc[df['Month'].isna(), 'Month'].fillna('December')

# Region_UNSD ##########
  # This might be Norther America or Central America, but since we cannot be sure,
  # we replace with 'Uncategorized'
df.loc[df['Region_UNSD'].isna(), 'Region_UNSD'] = df.loc[df['Region_UNSD'].isna(), 'Region_UNSD'].fillna('Uncategorized')

# Coordinates ##########
# Count NAs
print('\nNumber of coordinates missing ::::::::::::::::::::')
print(df.loc[df['Coordinates'].isna()].shape[0])
# Note 35 are in Pima County, 1 is in Off the coast of Libya
# For Pima County randomly pick coords from same Location
  # Set seed to 3 to avoid any NaN being picked and for reproducible results
random_coords = df.loc[df['Location']=='Pima Country Office of the Medical Examiner jurisdiction, Arizona, USA (see coordinates for exact location)',
                       'Coordinates'].sample(n=35, random_state=3).to_list()
print('\nRandomly picked close coordinates ::::::::::::::::::::')
print(random_coords)
# Replace
df.loc[(df['Location']=='Pima Country Office of the Medical Examiner jurisdiction, Arizona, USA (see coordinates for exact location)')
       & (df['Coordinates'].isna()), 'Coordinates'] = random_coords
# Remaining NAs
print('\nRemaining Coordinates NAs ::::::::::::::::::::')
print(df.loc[df['Coordinates'].isna()].shape[0])
# For Off the coast of Libya, we take a random place near similar locations
df.loc[(df['Location']=='off the coast of Libya')
       & (df['Coordinates'].isna()), 'Coordinates'] = '33.387319, 13.252497'
# Remaining NAs
print('\nRemaining Coordinates NAs ::::::::::::::::::::')
print(df.loc[df['Coordinates'].isna()].shape[0])

# Route ##########
# We simply replace NAs with 'Unknown'
df.loc[df['Route'].isna(), 'Route'] = df.loc[df['Route'].isna(), 'Route'].fillna('Unknown')

# Source ##########
# We simply replace NAs with 'Unknown'
df.loc[df['Source'].isna(), 'Source'] = df.loc[df['Source'].isna(), 'Source'].fillna('Unknown')

# Print final number of NAs
# Print summary
print('\nFinal number of sum of NAs ::::::::::::::::::::')
print(df.isna().sum().sum())
```

## 2.3 Save
